```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet("/content/train-00000-of-00001-b943ea66e0040b18.parquet")
```

```python
data_df = pd.DataFrame(df)
data_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

X = data_df['synopsis'].values
y = data_df['genre'].values
```

```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.regularizers import l2
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Tokenize input texts
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X)
X_encoded = tokenizer.texts_to_sequences(X)
X_padded = pad_sequences(X_encoded, padding='post')
```

```python
"""
vocab_size = len(tokenizer)
embedding_dim = 100
hidden_dim = 128
output_dim = len(label_encoder.classes_)
```

```python
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=128, input_length=X_padded.shape[1]))
model.add(LSTM(64, kernel_regularizer=l2(0.01), recurrent_regularizer=l2(0.01), bias_regularizer=l2(0.01), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(64, kernel_regularizer=l2(0.01), recurrent_regularizer=l2(0.01), bias_regularizer=l2(0.01), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(32, kernel_regularizer=l2(0.01), recurrent_regularizer=l2(0.01), bias_regularizer=l2(0.01)))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.01), bias_regularizer=l2(0.01)))
model.add(Dense(len(set(y_encoded)), activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
# X_np = encoded_inputs['input_ids'].numpy()
# y_np = encoded_labels
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_padded, y_encoded, test_size=0.2, random_state=42)
```

```python
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, batch_size=32)
```
